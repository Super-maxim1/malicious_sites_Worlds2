```python
import pandas as pd
import re
import tkinter as tk
from tkinter import messagebox


def check_safety(city_name):
    city_data = data_tg[data_tg['city'].str.lower() == city_name.lower()]

    if not city_data.empty:
        safety_level = city_data['danger_level'].values[0]
        response = f"Уровень безопасности города {city_name}: {safety_level}"
    else:
        response = f"Информации о безопасности города {city_name} нет в базе."

    return response

def help_command():
    return ("Привет! Я бот, основанный на CSV-файле. Вот как использовать:\n\n"
            "/help - Получить справку по командам\n"
            "Если я поеду в <город>, сколько опасных перекрёстков встретиться - Узнать уровень опасности города")

def process_command():
    user_input = entry.get()
    
    if user_input.lower() == '/help':
        result_text.set(help_command())
    else:
        command_pattern = re.compile(r"поеду в ([\w\s]+),.*опасных перекрёстков")
        match = command_pattern.search(user_input.lower())

        if match:
            city_name = match.group(1).strip()
            result = check_safety(city_name)
            result_text.set(result)
        else:
            result = check_safety(user_input)
            result_text.set(result)

root = tk.Tk()
root.title("Бот по безопасности городов")

tk.Label(root, text="Введите команду:").pack(pady=10)
entry = tk.Entry(root, width=50)
entry.pack(pady=10)

result_text = tk.StringVar()
result_text.set("Результат будет отображен здесь.")
result_label = tk.Label(root, textvariable=result_text)
result_label.pack(pady=10)

process_button = tk.Button(root, text="Выполнить", command=process_command)
process_button.pack(pady=10)

exit_button = tk.Button(root, text="Выход", command=root.destroy)
exit_button.pack(pady=10)

root.mainloop()
```

```python
!pip install python-telegram-bot==13.13
import pandas as pd
import re
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters

# Загрузка данных из CSV-файла
data = data_tg

TOKEN = '6887463107:AAFI7niqOU5OBPmADdBu6xQJlA70tRnuBl4' 
updater = Updater(token=TOKEN, use_context=True)
dispatcher = updater.dispatcher

# Обработчик команды /start
def start(update, context):
    context.bot.send_message(chat_id=update.effective_chat.id, text="Привет! Я бот, основанный на CSV-файле. Узнайте уровень опасности города, написав его название.")

# Обработчик текстовых сообщений
def check_safety(update, context):
    user_message = update.message.text
    city_name = extract_city_name(user_message)

    if city_name:
        city_data = data[data['city'].str.lower() == city_name.lower()]

        if not city_data.empty:
            safety_level = city_data['danger_level'].values[0]
            response = f"Уровень безопасности города {city_name}: {safety_level}"
        else:
            response = f"Информации о безопасности города {city_name} нет в базе."
    else:
        response = "Не удалось распознать команду. Пожалуйста, уточните запрос."

    context.bot.send_message(chat_id=update.effective_chat.id, text=response)

def extract_city_name(message):
    command_pattern = re.compile(r"поеду в ([\w\s]+),.*опасных перекрёстков")
    match = command_pattern.search(message.lower())

    if match:
        return match.group(1).strip()
    else:
        return message.strip()

# Обработчик команды /help
def help_command(update, context):
    help_text = "Привет! Я бот, основанный на CSV-файле. Вот как использовать:\n\n"
    help_text += "/start - Начать взаимодействие с ботом\n"
    help_text += "Если я поеду в <город>, сколько опасных перекрёстков встретиться - Узнать уровень опасности города\n"
    help_text += "<город> - Узнать уровень опасности города\n"
    help_text += "/cities - Показать список доступных городов\n"
    help_text += "/help - Получить справку по командам"
    context.bot.send_message(chat_id=update.effective_chat.id, text=help_text)

# Обработчик команды /cities
def list_cities(update, context):
    cities_list = "\n".join(data['city'])
    response = f"Список доступных городов:\n{cities_list}"
    context.bot.send_message(chat_id=update.effective_chat.id, text=response)

# Добавляем обработчики команд и текстовых сообщений
start_handler = CommandHandler('start', start)
check_safety_handler = MessageHandler(Filters.text & ~Filters.command, check_safety)
help_handler = CommandHandler('help', help_command)
cities_handler = CommandHandler('cities', list_cities)

dispatcher.add_handler(start_handler)
dispatcher.add_handler(check_safety_handler)
dispatcher.add_handler(help_handler)
dispatcher.add_handler(cities_handler)

# Запускаем бота
updater.start_polling()
updater.idle()
```
